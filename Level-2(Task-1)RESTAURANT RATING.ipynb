# ANALYZE THE DISTRIBUTION OF AGGREGATE RATINGS AND DETERMINE THE MOST COMMON RATING RANGE.

```python
df.columns
```

### Ploting the Distribution of Aggregate Ratings

```python
plt.hist(df['Aggregate rating'], bins = 5,color ='brown', edgecolor='black')
plt.xlabel("Aggregate Rating")
plt.ylabel("Frequency")
plt.title("Distribution of Aggregate Ratings")
#plt.grid(True)
plt.xticks(rotation=0)
plt.show()
```

### Determine the most common rating range


```python
rating_counts = df['Aggregate rating'].value_counts()

most_common_rating = rating_counts.idxmax()
most_common_rating_count = rating_counts.max()
print(f"The most common rating range is {most_common_rating}, with {most_common_rating_count} occurrences.")
```
