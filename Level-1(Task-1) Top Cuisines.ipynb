# DETERMINE THE TOP THREE MOST COMMON CUISINES IN THE DATASET.

```python
df['Cuisines'].value_counts()
```

```python
df['Cuisines'].value_counts().head(3)
```

```python
Top3_Cuisines = df['Cuisines'].value_counts().head(3)
```

```python
colours = ['green', 'blue', 'purple']
plt.bar(Top3_Cuisines.index, Top3_Cuisines.values, color = colours)
plt.title('Top Three Cuisines')
plt.xlabel('Cuisines')
plt.ylabel("Count")
plt.xticks(rotation =45)
plt.show()
```

# Calculate the percentage of restaurants that serve each of the top cuisines.

```python
total_restaurant = len(df)
print(total_restaurant)
top_cuisine10 =df['Cuisines'].value_counts().head()
percentages = (top_cuisine10 / total_restaurant) * 100
print("The Market share of Top 5 Cuisines are : ", percentages)
```

```python
plt.bar(top_cuisine10.index, percentages.values, color=colours)
plt.xlabel('Cuisine Name')
plt.ylabel('Percentage')
plt.title('Top five Cuisines %')
plt.xticks(rotation =45)
plt.figure(figsize=(12,6))
plt.show()
```
